```python
archivo = "/Users/heribertovs/Library/CloudStorage/OneDrive-ITESO/Semestre 6/Programacion para Mineria De Datos/proyecto_mineria_datos/data/raw/datos_2024.xlsx"

dfs = []
```

```python
import pandas as pd
import numpy as np

estaciones = {
    "AGU": "Aguilas",
    "ATM": "Atemajac",
    "CEN": "Centro",
    "COU": "Country",
    "PIN": "Las Pintas",
    "LDO": "Loma Dorada",
    "MIR": "Miravalle",
    "OBL": "Oblatos",
    "SAN": "Santa Fe",
    "SFE": "San Juan de Ocotán",
    "SMT": "Santa Margarita",
    "TLA": "Tlaquepaque",
    "VAL": "Vallarta"
}

flags = ["SE", "IF", "IO", "IR", "ND", "VE", "NE", "IC"]

for id_est, nombre in estaciones.items():
    
    df = pd.read_excel(archivo, sheet_name=id_est)

    # limpiar nombres de columnas
    df.columns = df.columns.str.strip()

    # reemplazar banderas
    df = df.replace(flags, np.nan)

    # convertir numéricos
    cols = ["O3", "PM10", "PM2.5"]
    for col in cols:
        df[col] = pd.to_numeric(df[col], errors="coerce")

    # datetime
    df["Fecha"] = pd.to_datetime(df["Fecha"]).dt.normalize()
    df["datetime"] = df["Fecha"] + pd.to_timedelta(df["Hora"], unit="h")

    # agregar estación
    df["id_estacion"] = id_est
    df["estacion"] = nombre

    # seleccionar columnas
    df = df[["id_estacion", "estacion", "datetime", "O3", "PM10", "PM2.5"]]

    dfs.append(df)
```

```python
df_final = pd.concat(dfs, ignore_index=True)
```
