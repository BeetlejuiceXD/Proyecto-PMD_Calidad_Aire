### **Exploracion inicial**

Se hizo una exploracion inicial de los datos crudos y se observo que el formato de los datos cambia mucho por años por lo que se decidio separar todos los años por sus estaciones con el procentaje de nulos para poder tomar decisiones de que datos se pueden usar.

```python
datasets_por_anio = {}

import pandas as pd
import numpy as np

flags = ["SE", "IF", "IO", "IR", "ND", "VE", "NE", "IC", "VZ"]

estaciones = {
    "AGU": "Aguilas",
    "ATM": "Atemajac",
    "CEN": "Centro",
    "COU": "Country",
    "PIN": "Las Pintas",
    "LDO": "Loma Dorada",
    "MIR": "Miravalle",
    "OBL": "Oblatos",
    "SAN": "Santa Fe",
    "SFE": "San Juan de Ocotán",
    "SMT": "Santa Margarita",
    "TLA": "Tlaquepaque",
    "VAL": "Vallarta"
}

# mapa inverso por si viene el nombre completo
nombre_a_id = {v: k for k, v in estaciones.items()}

def limpiar_anio(archivo_excel: str, anio: int) -> pd.DataFrame:
    columnas_finales = ["anio", "id_estacion", "estacion", "datetime", "O3", "PM10", "PM2.5"]

    try:
        excel = pd.ExcelFile(archivo_excel)
        hojas = excel.sheet_names
        print(f"{anio}: {hojas}")
    except Exception as e:
        print(f"No se pudo abrir {anio}: {e}")
        return pd.DataFrame(columns=columnas_finales)

    # -----------------------------
    # FORMATO 1: hoja Data
    # -----------------------------
    if "Data" in hojas:
        try:
            df = pd.read_excel(archivo_excel, sheet_name="Data")
            df.columns = df.columns.str.strip()
            df = df.replace(flags, np.nan)

            columnas_necesarias = ["STATION", "DATE", "HOUR", "O3", "PM10", "PM2.5"]
            faltantes = [c for c in columnas_necesarias if c not in df.columns]
            if faltantes:
                print(f"{anio}: faltan columnas formato Data {faltantes}")
                return pd.DataFrame(columns=columnas_finales)

            for col in ["O3", "PM10", "PM2.5"]:
                df[col] = pd.to_numeric(df[col], errors="coerce")

            df["datetime"] = pd.to_datetime(df["DATE"], errors="coerce")
            df["id_estacion"] = df["STATION"].astype(str).str.strip()
            df["estacion"] = df["id_estacion"].map(estaciones)
            df["anio"] = anio

            return df[columnas_finales]

        except Exception as e:
            print(f"Error procesando formato Data {anio}: {e}")
            return pd.DataFrame(columns=columnas_finales)

    # -----------------------------
    # FORMATO 2: hoja SIMAJ XXXX
    # -----------------------------
    hoja_simaj = next((h for h in hojas if "SIMAJ" in h.upper() or "BASE DATOS" in h.upper()), None)
    if hoja_simaj is not None:
        try:
            df = pd.read_excel(archivo_excel, sheet_name=hoja_simaj)
            df.columns = df.columns.str.strip()
            df = df.replace(flags, np.nan)

            columnas_necesarias = ["Estación", "Date_Time", "Hora", "O3", "PM10", "PM2.5"]
            faltantes = [c for c in columnas_necesarias if c not in df.columns]
            if faltantes:
                print(f"{anio}: faltan columnas formato SIMAJ {faltantes}")
                print(df.columns.tolist())
                return pd.DataFrame(columns=columnas_finales)

            for col in ["O3", "PM10", "PM2.5"]:
                df[col] = pd.to_numeric(df[col], errors="coerce")

            # Date_Time ya trae la fecha
            df["datetime"] = pd.to_datetime(df["Date_Time"], errors="coerce")

            # si Estación viene como nombre completo
            df["estacion"] = df["Estación"].astype(str).str.strip()

            # intentar obtener id
            # 1) si la columna ya viene con abreviatura, usa mapa inverso
            df["id_estacion"] = df["estacion"].map(nombre_a_id)

            # 2) si no encontró nada, quizá Estación ya trae código
            faltan_id = df["id_estacion"].isna()
            df.loc[faltan_id, "id_estacion"] = df.loc[faltan_id, "estacion"]

            # normalizar nombres si vino código
            df["estacion"] = df["id_estacion"].map(estaciones).fillna(df["estacion"])

            df["anio"] = anio

            return df[columnas_finales]

        except Exception as e:
            print(f"Error procesando formato SIMAJ {anio}: {e}")
            return pd.DataFrame(columns=columnas_finales)

    # -----------------------------
    # FORMATO 3: hoja por estación
    # -----------------------------
    dfs = []

    for id_estacion, nombre_estacion in estaciones.items():
        if id_estacion not in hojas:
            continue

        try:
            df = pd.read_excel(archivo_excel, sheet_name=id_estacion)
            df.columns = df.columns.str.strip()
            df = df.replace(flags, np.nan)

            columnas_necesarias = ["Fecha", "Hora", "O3", "PM10", "PM2.5"]
            faltantes = [c for c in columnas_necesarias if c not in df.columns]
            if faltantes:
                print(f"{anio} - {id_estacion}: faltan columnas {faltantes}")
                continue

            for col in ["O3", "PM10", "PM2.5"]:
                df[col] = pd.to_numeric(df[col], errors="coerce")

            df["Fecha"] = pd.to_datetime(df["Fecha"], errors="coerce").dt.normalize()
            df["datetime"] = df["Fecha"] + pd.to_timedelta(df["Hora"], unit="h")

            df["id_estacion"] = id_estacion
            df["estacion"] = nombre_estacion
            df["anio"] = anio

            df = df[columnas_finales]
            dfs.append(df)

        except Exception as e:
            print(f"Error en {anio} - {id_estacion}: {e}")

    if not dfs:
        return pd.DataFrame(columns=columnas_finales)

    return pd.concat(dfs, ignore_index=True)
```

```python
datasets_por_anio = {}

for anio, archivo in archivos_anuales.items():
    print(f"Procesando {anio}...")
    datasets_por_anio[anio] = limpiar_anio(archivo, anio)
```

```python
porcentajes_por_anio = {}

for anio, df_anio in datasets_por_anio.items():
    porcentajes_por_anio[anio] = (
        df_anio.groupby("estacion")[["O3", "PM10", "PM2.5"]]
        .apply(lambda x: x.isnull().mean() * 100)
        .round(2)
    )
```

```python
for anio in sorted(porcentajes_por_anio.keys()):
    print(f"\n========== AÑO {anio} ==========")
    print(porcentajes_por_anio[anio].to_string())

porcentajes_todos = pd.concat(
    [tabla.assign(anio=anio) for anio, tabla in porcentajes_por_anio.items()],
    axis=0
).reset_index()

porcentajes_todos = porcentajes_todos[["anio", "estacion", "O3", "PM10", "PM2.5"]]

porcentajes_todos
```
